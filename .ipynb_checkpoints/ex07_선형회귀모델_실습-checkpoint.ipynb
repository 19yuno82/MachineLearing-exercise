# 1.문제정의
- 보스턴 집값 데이터를 활용해서 집값을 예측해보자

# 2.데이터 수집

```python
from sklearn.datasets import fetch_openml
boston = fetch_openml('boston')
#분홍색 경고창 = 버전에 따라 다른 코드, 필수 파라미터 누락, 
```

```python
boston.keys()
```

```python
import pandas as pd
import numpy as np
```

```python
X=pd.DataFrame(boston['data'], columns=boston['feature_names'])
y=pd.DataFrame(boston['target'])
```

```python
X.info()
```

```python
#categroy > int 로 변경
X[['CHAS','RAD']]=X[['CHAS','RAD']].astype(int)
```

```python
X.info()
```

# 3.데이터 전처리

```python
#데이터 스케일링
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
sds=StandardScaler()
#스케일러에게 학습시키기(변환할 범위 정하기)
sds.fit(X)
#변환
X_sds=sds.transform(X)
```

# 4.탐색적 데이터 분석
- 생략

# 5.모델 선택 및 하이퍼파라미터 튜닝

#### Linear model 장점
- 결과예측(추론) 속도가 빠르다
- 대용량 데이터에도 충분히 활용 가능하다
- 특성이 많은 데이터 세트라면 훌륭한 성능을 낼 수 있다
#### Linear model 단점
- 특성이 적은 저차원 데이터에서는 다른 모델의 일반화 성능이 더 좋을 수 있다
    - 특성 확장을 하기도 한다
- LinearRegression는 복잡도를 제어할 방법이 없어 과대적합되기 쉽다
    - 모델 정규화를 통해 과대적합을 제어한다

```python
#train test 나누기
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, random_state= 3)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
#수학 공식을 이용한 모델, 하이퍼파라미터 굳이 쓸 게 없다
lr=LinearRegression()
#Linear Regression + 규제
#Lasso : 모든 가중치에 똑같은 값으로 규제 >  특성 선택
#Ridge : 모든 가중치에 똑같은 비율로 규제 > 모든 특성 사용
```

```python
#규제 조절 : alpha
#alpha 상승 : 규제 많이 가한다
#alpha 하강 : 규제를 조금 가한다 > 너무 낮으면 LinearRegression과 같은 결과
lasso=Lasso(alpha=0.1)
ridge=Ridge(alpha=0.1)
```

```python
#점진적으로 정답을 찾아가는 모델 : 경사하강법
from sklearn.linear_model import SGDRegressor
#max_iter : 가중치 업데이트 횟수
# -값 갱신 최대 횟수
# -최대횟수 도달 전에 최적의 값이라 생각하면 정지
#eta0 : 학습률 (0 ~ 1, 0.1↓)
# -수정을 얼마나 할건지
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=6,
 max_features=0.6,
 min_samples_leaf=10,
 min_samples_split=20,
 n_estimators=120)
```

# 6.학습

```python
lr.fit(X_train,y_train)
```

```python
lasso.fit(X_train,y_train)
```

```python
ridge.fit(X_train,y_train)
```

# 7.예측 및 평가
- 회귀평가지표 : R2 Score

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
score=cross_val_score(lr,X_train,y_train,cv=5)
score.mean()
```

```python
cross_val_score(lasso,X_train,y_train,cv=5).mean()
```

```python
cross_val_score(ridge,X_train,y_train,cv=5).mean()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_sds,y, random_state= 3)
```

```python
lr.fit(X_train,y_train)
score=cross_val_score(lr,X_train,y_train,cv=5)
score.mean()
```

```python
#GridSearch 사용하기
params={
    'n_estimators':[60,80,100,120,140],
    'max_depth':[3,4,5,6],
    'min_samples_split':[15,20,25,30],
    'min_samples_leaf':[10,15,20,25],
    'max_features' : [0.4,0.5,0.6,0.7]
}
grid_search=GridSearchCV(RandomForestRegressor(),params,cv=5)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```
