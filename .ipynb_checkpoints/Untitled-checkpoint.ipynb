# 3.데이터 전처리

```python
import pandas as pd
import numpy as np
```

```python
#문제데이터 분류
X=pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
X.head()
```

```python
#정답데이터
y=pd.DataFrame(cancer['target'], columns=['cancer'])
y.head()
```

```python
data=pd.concat([y,X], axis=1)
data.corr()
```

## 데이터 스케일링 사용하기

```python
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler
```

```python
X_StandardScaler=StandardScaler()
X_sds=X_StandardScaler.fit_transform(X)
```

```python
X_sds=pd.DataFrame(X_sds,index=X.index,columns=X.columns)
X_sds.head()
```

```python
X_RobustScaler=RobustScaler()
X_rbs=X_RobustScaler.fit_transform(X)
```

```python
X_rbs=pd.DataFrame(X_rbs,index=X.index,columns=X.columns)
X_rbs.head()
```

```python
X_MinMaxScaler=MinMaxScaler()
X_mms=X_MinMaxScaler.fit_transform(X)
```

```python
X_mms=pd.DataFrame(X_mms,index=X.index,columns=X.columns)
X_mms.head()
```
