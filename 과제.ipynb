# 2. 데이터 수집

```python
from sklearn.datasets import fetch_openml
```

```python
boston = fetch_openml('boston')
```

```python
boston.keys()
```

# 3. 데이터 전처리

```python
import pandas as pd

df_X = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
df_boston = pd.concat([df_X,df_y], axis=1)

df_boston.head(5)
```

```python
df_X.info()
```

```python
# X(문제) 데이터 y(정답) 데이터 만들기
df_X['CHAS'] = df_X['CHAS'].astype(int)
df_X['RAD'] = df_X['RAD'].astype(int)
```

```python
df_y=df_y.astype(int)
```

```python
# 데이터 스케일링 하기
from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler
rbs=RobustScaler()
sds=StandardScaler()
mms=MinMaxScaler()
```

```python
X_rbs=rbs.fit_transform(df_X)
X_sds=sds.fit_transform(df_X)
X_mms=mms.fit_transform(df_X)
```

# 5. 모델 선택 및 하이퍼 파라미터 튜닝

```python
# X_train,X_test,y_train,y_test 만들기
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_sds,df_y,test_size=0.3,random_state=43)
```

```python
# 앙상블 모델 불러오기
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
pre_y=rf.predict(X_test)
metrics.accuracy_score(pre_y, y_test)
```

```python
ab=AdaBoostClassifier()
ab.fit(X_train,y_train)
pre_y=ab.predict(X_test)
metrics.accuracy_score(pre_y,y_test)
```

```python
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
pre_y=gb.predict(X_test)
metrics.accuracy_score(pre_y,y_test)
```

```python
# GridSearch 사용해서 최적의 하이퍼 파라미터 찾기
from sklearn.model_selection import GridSearchCV
param={
    'n_estimators':[30,50,70,100],
    'max_depth':[2,3,4,5],
    'min_samples_split':[2,5,10,15],
    'min_samples_leaf':[2,3,4,5]
}
grid_model=GridSearchCV(RandomForestClassifier(),param,cv=5)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```
